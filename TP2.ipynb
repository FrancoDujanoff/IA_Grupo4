## Ejercicios Teóricos

1. ¿Qué diferencia hay entre una estrategia de búsqueda Informada y una estrategia de búsqueda No Informada?

En una busqueda no informada o tambien llamada a ciegas, no se cuenta con ningun tipo de informacion adicional sobre los estados que no son objetivos, por otro lado la estragia de busqueda informada o tambien llamada heuristica cuenta con estrategias que le permiten saber si un estado objetivo es mas prometedor que otro.
Es decir que en la no informada, no se dispondra de ninguna pista sobre que tan cerca esta cada estado de la meta, por eso se explora el espacio sin guia mas alla de la estructura. En cambio en la informada, se usa un conocimiento adicional que estima el costo de la ruta mas barata desde el estado n hasta la meta, esta pista se utiliza para dirigirse hacia los estados mas prometedores 

2. ¿Qué es una heurística y para qué sirve?

Una heuristica es una funcion h(n) que dado cierto nodo en el espacio, devuelve una estimacion del costo minimo restante para alcanzar un estado meta desde n, no garantiza exactitud pero orienta la busqueda. Sirve para guiar la busqueda, reducir costo computacional, mejorar eficiencia, balancear precision y rapidez, etc.

3. ¿Es posible que un algoritmo de búsqueda no tenga solución?

Si es posible que un algoritmo de busqueda no tenga solucion, puede deberse a que el problema en si no tiene solucion o que se haya definido incorrectamente el problema  o que el algoritmo tenga limitaciones o que los recursos sean insuficientes.

Como por ejemplo en la Busqueda de Profundidad Limitada, si definimos un limite de profunidad menor al necesario, no llegara a ninguna solucion, o tambien en la Busqueda A*, como guarda en memoria todos los nodos generados, se puede llegar a ocupar toda la memoria disponible antes de hallar una solucion.

```python
import numpy as np
import matplotlib.pyplot as plt
from heapq import heappush, heappop

def manhattan(a, b):
    return abs(a[0] - b[0]) + abs(a[1] - b[1])

def vecinos4(r,c, N):
    """Vecinos arriba/abajo/izquierda/derecha dentro del tablero."""
    for dr, dc in [(-1,0), (1,0), (0,-1), (0,1)]:
        nr, nc = r + dr, c + dc
        if 0 <= nr < N and 0 <= nc < N:
            yield (nr, nc)

def astar(grilla, inicio_rc, fin_rc):
    """
    A* en grilla 0/1 (0=libre, 1=pared). Movimiento 4-dir con costo 1.
    Retorna la lista de celdas (r,c) desde inicio a fin (incluidos),
    o None si no hay camino.
    """
    N = grilla.shape[0]
    start, goal = inicio_rc, fin_rc

    # open set como heap de (f, g, nodo)
    openh = []
    heappush(openh, (manhattan(start, goal), 0, start))

    came_from = {}          # para reconstruir
    gscore = {start: 0}     # mejor g conocido para cada nodo
    cerrado = set()         # visitados/expandidos

    while openh:
        f, g, actual = heappop(openh)
        if actual in cerrado:
            continue
        if actual == goal:
            # reconstruir camino
            path = [actual]
            while actual in came_from:
                actual = came_from[actual]
                path.append(actual)
            path.reverse()
            return path

        cerrado.add(actual)
        r, c = actual
        for nr, nc in vecinos4(r, c, N):
            if grilla[nr, nc] == 1:        # pared => no se puede
                continue
            nuevo_g = gscore[actual] + 1   # costo unitario
            if (nr, nc) not in gscore or nuevo_g < gscore[(nr, nc)]:
                gscore[(nr, nc)] = nuevo_g
                came_from[(nr, nc)] = actual
                fscore = nuevo_g + manhattan((nr, nc), goal)
                heappush(openh, (fscore, nuevo_g, (nr, nc)))
    return None


def leer_entero(mensaje, minimo, maximo):
    #Esta funcion va a pedir un entero por teclado que este entre un minimo y un maximo
    while True:
        try:
            v = int(input(mensaje))
            if minimo <= v <= maximo:
                return v
            print("Valor fuera de rango")
        except ValueError:
            print("Ingrese un numero entero")

def pedir_y_marcar_inicio_fin(grilla, ax):
    N = grilla.shape[0]
    print("\nCoordenadas (x,y) en 0..17, origen arriba-izquierda.")
    xi = 3
    yi = 3
    xf = 15
    yf = 15
    
    inicio = (yi, xi)
    fin = (yf, xf)
    
    # Aviso si cayó sobre una pared (no lo bloqueo todavía)
    if grilla[inicio] == 1:
        print("⚠️  Atención: el INICIO cae sobre una pared (se marcará igual).")
    if grilla[fin] == 1:
        print("⚠️  Atención: el FIN cae sobre una pared (se marcará igual).")
    
    #Marcamos los puntos
    # Creamos una "capa" del tamaño de la grilla con transparencia (RGBA)
    mask = np.zeros((N, N, 4), dtype=float)   # todo transparente

    # Colores RGBA: [R,G,B,Alpha]. Alpha=1 => opaco
    mask[yi, xi] = [0.2, 0.4, 1.0, 1.0]       # inicio azul en (fila=yi, col=xi)
    mask[yf, xf] = [1.0, 0.0, 0.0, 1.0]       # fin rojo   en (fila=yf, col=xf)

    # Superponer la capa de colores sobre la grilla
    ax.imshow(mask, origin='upper', interpolation='nearest', zorder=3)

    # (Opcional) leyenda sin dibujar puntos extra
    ax.plot([], [], 's', color='blue',  markersize=10, label='Inicio')
    ax.plot([], [], 's', color='red',   markersize=10, label='Fin')
    ax.legend(loc='upper left')
    

    return inicio, fin
    
def dibujar_camino(ax, path, N, color=(1.0, 1.0, 0.0, 0.9), zorder=2):
    """
    Pinta el camino sobre una máscara RGBA del mismo tamaño que la grilla N×N.
    """
    if not path:
        return

    mask = np.zeros((N, N, 4), dtype=float)
    for r, c in path:
        mask[r, c] = color

    # Guardar/restaurar límites para que imshow no reescale
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()
    ax.imshow(mask, origin='upper', interpolation='nearest', zorder=zorder)
    ax.set_xlim(xlim)
    ax.set_ylim(ylim) 

def main():
    N=18 #Defino el tamano de la grilla
    grilla = np.zeros((N,N), dtype=int) #Creamos una matriz para representar cada posicion de la grilla, 0 sera libre y 1 sera pared

    fig, ax = plt.subplots(figsize=(6,6)) #Creamos la figura y el eje de dibujo

    pared = [
        (2, 11), (3, 10),
        (4, 11), (5, 11), (6, 11), (7, 11), (8, 11),
        (9, 11), (10, 11), (11, 11), (11, 12),
        (11, 13), (11, 14), (11, 15)
    ]

    #Armamos la pared es importante que la primer coordenada es Y y la segunda es X
    for (r, c) in pared:
        grilla[r, c] = 1

    ax.imshow(grilla, origin='upper', interpolation='nearest',
            cmap='gray_r', vmin=0, vmax=1)

    #Marcamos lineas de la cuadricula cada 1 celda
    #Las 'minor ticks' son marcas intermedias donde apoyamos las lineas del grid
    ax.set_xticks(np.arange(-0.5, N, 1), minor=True) #Lineas verticales
    ax.set_yticks(np.arange(-0.5, N, 1), minor=True) #Lineas horizonales

    ax.grid(which='minor', color='lightgray', linestyle='-', linewidth=0.8) #Activamos las lineas en esas marcas menores

    #Quitamos las marcas y numeros de los ojes
    ax.set_xticks([])
    ax.set_yticks([])

    #Titulo
    ax.set_title("Cuadricula 18x18")
    
    #Pedimos y marcamos inicio/fin
    inicio_rc, fin_rc = pedir_y_marcar_inicio_fin(grilla, ax)

    #Ajustamos los margenes para que no se corte nada
    plt.tight_layout()
    
    camino = astar(grilla, inicio_rc, fin_rc)
    if camino is None:
        print("No hay camino 🤷 (la pared bloquea la ruta entre inicio y fin).")
    else:
        camino_interno = [p for p in camino if p not in (inicio_rc, fin_rc)]
        dibujar_camino(ax, camino_interno, N=grilla.shape[0]) 

    #Mostramos la ventana con la cuadricula
    plt.show()
    
#Punto de entrada
if __name__ == "__main__":
    main()





```
